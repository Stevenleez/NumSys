```python
import matplotlib.pyplot as plt

def my_lin_interp(x, y, X):
    Y = []
    
    for Xs in X:
        i = 0
        while i < len(x) and x[i] < Xs:
            i += 1
        
        if i < len(x):
            linInt = y[i-1] + ((y[i] - y[i-1]) * (Xs - x[i-1])) / (x[i] - x[i-1])
            Y.append(linInt)
            
        else:
            Y.append("n/a")
    
    return Y
        

x = [-2, 0, 2, 3, 6]
y = [2, 0, 2, 1, 2]
X = [-1.0, -.5, .5, 1.0, 2.5, 4, 5]

Y = my_lin_interp(x, y, X);
print(Y)

plt.plot(x, y, 'bo', label='Data Points')

# Plot for linear interpolation
plt.plot(X, Y, 'r-', label='Linear Interpolation')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Interpolation')
plt.legend()
plt.grid(True)
plt.show()
```

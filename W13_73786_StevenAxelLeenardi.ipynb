```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-poster')
%matplotlib inline

n = 4
h = (1 - 0) / n

# Get A
A = np.zeros((n+1, n+1))
A[0, 0] = 1
A[n, n] = 1
for i in range (1, n):
    A[i, i-1] = 1
    A[i, i] = -2.25
    A[i, i+1] = 1
    
print(A)

# Get b
b = np.zeros(n+1)
b[0] = 1.1752
b[-1] = 10.0179
print(b)

# Solve the linear equation
y = np.linalg.solve(A, b)
t = np.linspace(0, 1, 5)

result = np.column_stack((t, y))
print("Results:\n", result)

plt.figure(figsize=(10, 8))
plt.plot(t, y)
plt.plot(1, 10.0179, 'ro')
plt.xlabel('time(s)')
plt.ylabel('altitude (m)')
plt.show()
```

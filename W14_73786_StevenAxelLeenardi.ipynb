```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-poster')
%matplotlib inline

# sampling rate
sr = [5, 10, 20, 50, 100]
# sampling interval
ts = [1.0/s for s in sr]
# time vector
t = [np.arange(0, 2, interval) for interval in ts]

# frequency of the signal
freq1 = 5
freq2 = 2
x = [3 * np.sin(2 * np.pi * freq1 * t[i] + 3) + 2 * np.sin(2 * np.pi * freq2 * t[i] - 2) for i in range(len(sr))]

# Plot 1
plt.figure(figsize=(8, 8))
plt.subplot(211)
plt.plot(t[4], x[4], 'r')
plt.ylabel('Amplitude')
plt.xlabel('Time (s)')

# Plot 2
plt.figure(figsize=(14, 13))

for i in range(len(sr)):
    plt.subplot(len(sr), 1, i+1)
    plt.plot(t[i], x[i], 'r')
    plt.ylabel('Amplitude')
    plt.xlabel('Time (s)')
    plt.title(f'Sampling Rate: {sr[i]} Hz', style='italic')

plt.tight_layout()
plt.show()
```
